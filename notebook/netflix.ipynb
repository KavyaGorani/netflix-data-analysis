```python
movies_duration = df[df['type'] == 'Movie'].copy()

# extract numbers
movies_duration['duration'] = movies_duration['duration'].str.split().str[0]

# drop missing values
movies_duration = movies_duration.dropna(subset=['duration'])

# convert to int
movies_duration['duration'] = movies_duration['duration'].astype(int)

# find longest movie
longest_movie = movies_duration.sort_values(by='duration', ascending=False).head(10)

print(longest_movie[['title', 'duration']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x = df['type'] ,color="orange")
```

```python
df = df.dropna(subset=['country'])
```

```python
df['country'] = df['country'].str.split(',')
df = df.explode('country')
```

```python
df['country'] = df['country'].str.strip()
```

```python
top_countries = df['country'].value_counts().head(10)
```

```python
top_countries
```

```python
top_countries.plot(kind='bar')

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
```

```python
year_counts = df['year_added'].value_counts().sort_index()
```

```python
year_counts
```

```python
year_counts.plot(kind='line', marker='o')

plt.title("Netflix Content Growth Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```
